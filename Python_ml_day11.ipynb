## Accessor operators in Pandas--- .loc(), iloc()

```python
# .iloc -- based on indexing position --- 0,1,2---rows first, column second
## 0 row , all columns
print(house_data.iloc[0])
```

```python
## all row, 0 columns
house_data.iloc[:, 0]
```

```python
## range.. first 3 rows in 0 col
house_data.iloc[:3, 0]
```

```python
## can also pass as a list
house_data.iloc[[1,2,3],0]
```

```python
## negative indexing
house_data.iloc[-5:]
```

```python
### .loc() -- label based selection
house_data.loc[0, 'Suburb']
```

```python
## more than one column
print(house_data.loc[:5, ['Suburb', 'Address', 'Method', 'Date']])
```

### diff. bw .loc and .iloc in indexing - in .loc - 0:10.. 10 is inclusive, also works with strings // in .iloc.. 0:10 10 is exclusive

## Conditional Selection 

```python
## checking house in Suburb -- Abbotsford
house_data.Suburb == 'Abbotsford'
```

```python
## getting all data of Suburb -- Abbotsford
house_data.loc[house_data.Suburb == 'Abbotsford']
```

```python
## can use two conditions on data with ampersand &..suburb == 'Abbotsford' and Type == 'h'
house_data.loc[(house_data.Suburb == 'Abbotsford') & (house_data.Type == 'h')]
```

```python
## | Pipe .. as an OR cond. in Pandas.. house in Abbotsford Suburb or price less than 1230000
house_data.loc[(house_data.Suburb== 'Abbotsford') | (house_data.Price <= 1230000)]
```

## .isin 

```python
## .isin -- isin is lets you select data whose value "is in" a list of value
```

```python
house_data.loc[house_data.Suburb.isin(['Abbotsford', 'Williamstown'])]
```

## isnull -- for checking null values or data....notnull -- opposite of isnull..These methods let you highlight values which are (or are not) empty (NaN).

```python
house_data.loc[house_data.Landsize.isnull()]
```
